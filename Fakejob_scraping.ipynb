```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from datetime import datetime
```

```python
page =  requests.get('https://realpython.github.io/fake-jobs/')
```

```python
soup = BeautifulSoup(page.content,'html.parser')
```

```python
print(soup)
```

```python
print(soup.find_all("a"))
```

```python
fake = soup.find(class_='section')
print(fake)
```

```python
items = soup.find_all(class_='card-content')
print(items)
```

```python
print(items[0])
```

```python
print(items[0].find('h2',class_='title is-5').get_text())
print(items[0].find('h3',class_='subtitle is-6 company').get_text())
print(items[0].find(class_='location').get_text())


```

```python
Post = [item.find(class_='title is-5').get_text() for item in items]
Title = [item.find(class_='subtitle is-6 company').get_text() for item in items]
Location = [item.find(class_='location').get_text() for item in items]

print(Post)
print(Title)
print(Location)
```

```python
jobs = pd .DataFrame(
    {
        'Post': Post,
        'Title': Title,
        'Location': Location,
     })
print(jobs)
```

```python
jobs.to_csv('fake_jobs.csv')
```
